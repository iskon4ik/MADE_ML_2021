## Часть 1

### 1.1
Пусть А - событие виновности обвиняемого, B - событие совпадения группы крови с группой убийцы <br>
В начале прокурор вполне прав и оценивает вероятность B при условии не A: <br>
$$
P(B|\neg{A}) = \frac{P(\neg{A}|B)P(B)}{P(\neg{A})} \approx 0.01
$$
Но дальше он заявляет, что из этого следует, что обвиняемый виновен с вероятностью 0.99, т.е. он оценил
$$
P(B|A) = 0.99
$$
<br>
Если же он действительно хотел понять веротяность виновности обвиняегомого, он бы посчитал вероятность:
$$
P(A|B) = \frac{P(B|A)P(A)}{P(B)}
$$
которая будет сильно зависеть от общего населения города, т.к.
$$
P(A) = \frac{1}{количество\ жителей\ города}
$$
По сути, если в городе живет много человек и никаких других доказательств на обвиняемого, кроме совпадающей группы крови нет, то это лишь немного увеличивает вероятность его виновности, относительно других людей (возможно, в тесте есть вероятность ошибки)

### 1.2
Заявление адвоката верно, если только никаких других доказательств против обвиняемого нет. Если же, например, C - событие доступа к месту преступления (что обвиняемый изначально входил в группу подозреваемых по каким-то более конкретным уликам), то можно было бы оценить вероятность:
$$
P(A|BC) = \frac{P(BC|A)P(A)}{P(BC)}
$$
Например, если окажется, что было довольно мало людей, имеющих доступ к месту преступления, и только у этого обвиняемого совпала группа крови, то это будет довольно сильным доказательством, которое ни в коем случае нельзя откидывать.

### 1.3

$$
P(A|B) = \frac{P(B|A)P(A)}{P(B)} = 
\frac{P(B|A)P(A)}{P(B|A)P(A) + P(B|\neg{A})P(\neg{A})} = 
\frac{(1 - P(B|\neg{A}))\ P(A)}{(1 - P(B|\neg{A}))\ P(A) + P(B|\neg{A})P(\neg{A})}
$$

$$
P(A|\neg{B}) = \frac{P(\neg{B}|A)P(A)}{P(\neg{B})} = \frac{P(\neg{B}|A)P(A)}{P(\neg{B}|A)P(A) + P(\neg{B}|\neg{A})P(\neg{A})} + \frac{P(\neg{B}|A)P(A)}{P(\neg{B}|A)P(A) + (1 - P(\neg{B}|A))\ P(\neg{A})}
$$

$$
false\ positive\ prob = P(B|\neg{A})\\
false\ negative\ prob = P(\neg{B}|A)
$$

```python
false_positive_prob = 0.05
false_negative_prob = 0.05

disease_rate = 0.01

def prob_sick_with_test_pos(false_pos_prob, disease_rate):
    tmp1 = (1 - false_pos_prob) * disease_rate
    tmp2 = false_pos_prob * (1 - disease_rate)
    return tmp1 / (tmp1 + tmp2)

def prob_sick_with_test_neg(false_neg_prob, disease_rate):
    tmp1 = false_neg_prob * disease_rate
#     tmp2 = (1 - false_neg_prob) * (1 - disease_rate)
    tmp2 = (1 - false_positive_prob) * (1 - disease_rate)
    return tmp1 / (tmp1 + tmp2)
```

```python
prob_sick_with_test_pos(false_positive_prob, disease_rate)
```

```python
prob_sick_with_test_neg(false_negative_prob, disease_rate)
```

```python
arr_false_pos = np.arange(0.05, -0.0005, -0.0005)
arr_false_neg = np.arange(0.05, -0.0005, -0.0005)
```

```python
res = [prob_sick_with_test_pos(x, disease_rate) for x in arr_false_pos]
plt.xlabel('False positive proba')
plt.ylabel('Proba of disease with positive test')
plt.plot(arr_false_pos, res)
```

```python
res = [prob_sick_with_test_neg(x, disease_rate) for x in arr_false_neg]
plt.xlabel('False negative proba')
plt.ylabel('Proba of disease with negative test')
plt.plot(arr_false_neg, res)
```

Я бы порекоммендовал уменьшать вероятность ложно положительного исхода, т.к. с ним, даже если не получится убрать ее полностью, получится добиться гораздо большего увеличения вероятности.
